```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data(train=True):
  """
  Load binary datasets into their dictionary
  """
  if train:
    loc = './train/train_data.pickle'
  else:
    loc = './test/test_data.pickle'

  with open(loc, 'rb') as f:
    data = pickle.load(f)
    return data
  
list_dfs = load_data()
print(f'Length of dataset: {len(list_dfs)}')
```

```python
def plot_10_price(list_dfs, keys_to_plot, figheight=8, figwidth=36):
  colours = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan']

  fig, ((ax1, ax2, ax3, ax4, ax5), (ax6, ax7, ax8, ax9, ax10)) = plt.subplots(2, 5)
  fig.set_figheight(figheight)
  fig.set_figwidth(figwidth)

  axes = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9, ax10]

  for idx, axis in enumerate(axes):
    data = list_dfs[keys_to_plot[idx]]['close']

    axis.plot(range(len(data)), data, colours[idx])

ten_keys = list(list_dfs.keys())[0:10]
plot_10_price(list_dfs, ten_keys)
```
